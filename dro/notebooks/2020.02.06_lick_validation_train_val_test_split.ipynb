```python
import pandas as pd
import os
from tqdm import tqdm
import shutil
```

```python
base_data_path = r'//allen/programs/braintv/workgroups/nc-ophys/visual_behavior/lick_detection_validation'
folder_names = [
    'noncontact_frames',
    'tongue_contact_frames',
    'paw_contact_frames',
]
base_data_path
```

```python
data = []
for folder_name in folder_names:
    df = pd.DataFrame({'filename':os.listdir(os.path.join(base_data_path,folder_name))})
    df['folder'] = folder_name
    data.append(df)
data = pd.concat(data)
```

```python
data = data.sample(len(data)).reset_index()
data['set'] = None
data['label'] = None
```

```python
def label(folder):
    if folder.startswith('tongue'):
        return 'tongue'
    elif folder.startswith('paw'):
        return 'paw'
    else:
        return 'no_contact'
```

```python
data.loc[:0.8*len(data),'set'] = 'train'
data.loc[0.8*len(data):0.9*len(data),'set'] = 'validate'
data.loc[0.9*len(data):,'set'] = 'test'
data['label'] = data['folder'].map(label)
```

```python
data_dst = '/ssd1/lick_detection_model/three_class_model'
for idx,row in tqdm(data.iterrows()):
    src = os.path.join(base_data_path,row['folder'],row['filename'])
    dst = os.path.join(data_dst,row['set'],str(row['label']))
    if os.path.exists(dst) == False:
        os.mkdir(dst)
    shutil.copy(src,dst)
```
