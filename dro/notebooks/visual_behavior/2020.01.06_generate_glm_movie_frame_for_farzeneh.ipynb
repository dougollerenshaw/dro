```python
from visual_behavior_glm.glm import GLM
from visual_behavior.plotting import save_figure
```

```python
oeid = 1008408502 # vip example
version = '10a_L2_optimize_by_session'
glm = GLM(
    oeid, 
    version, 
    use_previous_fit=True,  # if True, uses cached fit if available (False by default)
    log_results=True, # if True, logs fit results to mongo database (True by default)
    log_weights=True, # if True, logs weights to mongo database (True by default)
    NO_DROPOUTS=False, # if True, does not perform dropout analysis (False by default)
    TESTING=False, # if True, fits only the first 6 cells in the experiment (False by default)
)
```

```python
cell_specimen_ids = glm.session.cell_specimen_table.index
```

```python
csid = int(cell_specimen_ids[1])
movie = glm.movie(csid, action='display_first_frame', start_frame=25050, end_frame=30000)
```

```python
save_figure(
    movie.fig, 
    '/allen/programs/braintv/workgroups/nc-ophys/visual_behavior/shared_figures/glm_video_frames_oeid={}_csid={}'.format(oeid, csid),
    formats=['.png','.pdf'],
    size=(24, 14),
)
```
