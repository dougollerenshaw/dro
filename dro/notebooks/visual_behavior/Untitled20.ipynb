```python
from visual_behavior.data_access import loading
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%widescreen
%matplotlib notebook
```

```python
def plot_dff_and_events(oeid, csid):
    
    dataset = loading.get_ophys_dataset(oeid)

    fig,ax = plt.subplots(3,1,figsize=(15,5), sharex=True)
    ax[0].plot(
        dataset.ophys_timestamps,
        dataset.dff_traces.loc[csid]['dff']
    )
    ax[1].plot(
        dataset.ophys_timestamps,
        dataset.events.loc[csid]['events']
    )
    ax[2].plot(
        dataset.ophys_timestamps,
        dataset.events.loc[csid]['filtered_events']
    )

    ax[0].set_title('dff')
    ax[1].set_title('events')
    ax[2].set_title('filtered events')

    fig.tight_layout()
    plt.subplots_adjust(top=0.9)
    fig.suptitle('ophys_experiment_id = {}, cell_specimen_id = {}'.format(oeid, csid))
    
    return fig, ax
```

```python
oeid = 796308505
csid = 1065546324
fig, ax = plot_dff_and_events(oeid, csid)
ax[1].set_xlim(2985, 3045)
```

```python
oeid = 796308505
dataset = loading.get_ophys_dataset(oeid)
for csid in dataset.cell_specimen_ids:
    plot_dff_and_events(oeid, csid)
```
