```python
import matplotlib.pyplot as plt
from allensdk.brain_observatory.behavior.behavior_ophys_session import BehaviorOphysSession

def make_plot(oeid, csid):
    
    sdk_session = BehaviorOphysSession.from_lims(oeid)

    fig,ax=plt.subplots(1,3,sharey=True,figsize=(15,7.5))

    cell = sdk_session.cell_specimen_table.loc[csid]

    max_projection = sdk_session.get_max_projection()
    ax[0].imshow(max_projection)
    ax[0].set_title('max_projection')

    ax[1].imshow(sdk_session.get_segmentation_mask_image())
    ax[1].set_title('SDK segmentation mask image')

    sdk_roi_mask = sdk_session.get_roi_masks().loc[csid].values
    ax[2].imshow(sdk_roi_mask)
    ax[2].set_title('SDK ROI mask')

    for axis in ax.flatten():
        axis.axvline(cell['x'])
        axis.axhline(cell['y'])

    fig.tight_layout()
```

```python
oeid = 1043833869
csid = 1044470301
make_plot(oeid, csid)
```
