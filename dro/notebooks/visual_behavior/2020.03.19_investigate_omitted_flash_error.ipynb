```python
summary = db.get_behavior_session_summary()
def make_int(v):
    if pd.notnull(v):
        return int(v)
    else:
        return None
summary['lims_id'] = summary['lims_id'].map(lambda v:make_int(v))
```

```python
np.sort(summary.rig_id.unique())
```

```python
ophys_summary = summary[
    summary['rig_id'].isin(['2P3','2P4','2P'])
]
print(len(ophys_summary))
ophys_summary.head()
```

```python
ii = 0
for idx,row in ophys_summary.iterrows():
    uuid = row['behavior_session_uuid']
    try:
        omitted_stimuli = db.get_behavior_data('omitted_stimuli',uuid)
        metadata = db.get_behavior_data('metadata',uuid)
    except:
        print('could not extract behavior data for uuid {}'.format(uuid))
    
    if len(omitted_stimuli) > 1:
        ophys_summary.at[idx,'omitted_flash_on_frame_0'] = omitted_stimuli.iloc[0]['frame'] == 0
        ophys_summary.at[idx,'omitted_flashes_exist'] = True
    else:
        ophys_summary.at[idx,'omitted_flashes_exist'] = False
        
    ophys_summary.at[idx,'camstim_version'] = metadata['platform_info']['camstim']
    
        
    ii += 1
    print('ii = {}'.format(ii), end = '\r')
```

```python
of_interest = ophys_summary[~pd.isnull(ophys_summary['omitted_flashes_exist']==True)]
cols = ['behavior_session_uuid','lims_id','omitted_flashes_exist','omitted_flash_on_frame_0','rig_id','stage','startdatetime','camstim_version']
error_summary  = of_interest[of_interest['omitted_flashes_exist']==True][cols].sort_values(by=['startdatetime'],ascending=True)
error_summary.to_csv('/home/dougo/omitted_flash_error_log.csv',index=False)
error_summary
```

```python
error_summary.query("omitted_flash_on_frame_0 == True")
```

```python
def get_95_percent_confidence_interval(ns,n):
    p_hat = ns/n
    lower_bound = p_hat - 1.96*np.sqrt((p_hat*(1-p_hat))/n)
    upper_bound = p_hat + 1.96*np.sqrt((p_hat*(1-p_hat))/n)
    return lower_bound,upper_bound
```

```python
print('the earliest occurrence of the error was on: {}'.format(error_summary.query("omitted_flash_on_frame_0 == True")['startdatetime'].min()))
print('\nthe most recent occurrence of the error was on: {}'.format(error_summary.query("omitted_flash_on_frame_0 == True")['startdatetime'].max()))

n_errors = error_summary['omitted_flash_on_frame_0'].astype(int).sum()
n_sessions = len(error_summary)
print('\nof {} total sessions, there have been {} occurrences of the error'.format(n_sessions,n_errors))

confidence_range = get_95_percent_confidence_interval(n_errors,n_sessions)
print('\nthat corresponds to an error fraction of {:0.4f}, with 95% confidence intervals spanning {:0.4f} to {:0.4f}'.format(n_errors/n_sessions,confidence_range[0],confidence_range[1]))

print('\nthe breakdown across rigs is\n{}'.format(error_summary[error_summary['omitted_flash_on_frame_0']==True]['rig_id'].value_counts()))

print('\nthe breakdown across camstim versions is\n{}'.format(error_summary[error_summary['omitted_flash_on_frame_0']==True]['camstim_version'].value_counts()))

print('\nthe breakdown across stages is\n{}'.format(error_summary[error_summary['omitted_flash_on_frame_0']==True]['stage'].value_counts()))
```
