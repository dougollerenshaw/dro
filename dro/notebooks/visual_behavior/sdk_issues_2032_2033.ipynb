```python
import allensdk.brain_observatory.behavior.behavior_project_cache as bpc
import matplotlib.pyplot as plt

my_cache_dir = '/allen/programs/braintv/workgroups/nc-ophys/Doug/allensdk_visualbehavior_cache'

bc = bpc.VisualBehaviorOphysProjectCache.from_s3_cache(cache_dir=my_cache_dir)
          
behavior_session_id = 847874940

dataset = bc.get_behavior_session(behavior_session_id)


help(dataset.get_rolling_performance_df)
```

```python
%matplotlib notebook
```

```python
rpdf = dataset.get_rolling_performance_df()

fig, ax = plt.subplots(3, 1, figsize = (15,5), sharex=True)
ax[0].plot(
    rpdf['hit_rate'],
    color='darkgreen'
)

ax[0].plot(
    rpdf['false_alarm_rate'],
    color='darkred'
)

ax[0].legend(['rolling hit rate', 'rolling false alarm rate'])

ax[1].plot(
    rpdf['rolling_dprime'],
    color='black'
)

ax[1].set_xlabel('trial number')
ax[0].set_ylabel('response rate')
ax[0].set_title('hit and false alarm rates')
ax[1].set_title("d'")

ax[2].plot(
    rpdf['reward_rate']
)
ax[2].plot(dataset.get_reward_rate(), linestyle = ':')
fig.tight_layout()
```
