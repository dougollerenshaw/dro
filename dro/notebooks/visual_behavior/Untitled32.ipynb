```python
import visual_behavior.database as db
import pandas as pd

error_string = 'the number of rising edges in the draw log'
query = {
    "traceback": {
        "$regex": error_string,
        "$options" :'i' # case-insensitive
    },
    "sdk_version": "2.4.0"
}
conn = db.Database('visual_behavior_data')
matching_errors = pd.DataFrame(list(conn['sdk_validation']['error_logs'].find(query))).drop(columns='_id')
conn.close()

print("{} behavior sessions are affected by this error".format(len(matching_errors['behavior_session_id'].unique())))
matching_errors.sample(10)
```
