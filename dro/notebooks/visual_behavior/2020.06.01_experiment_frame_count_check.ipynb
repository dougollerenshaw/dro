```python
list_of_dicts = []
for oeid in df.ophys_experiment_id.unique():
    _dict = {'ophys_experiment_id':oeid}
    dft = df.query('ophys_experiment_id == {}'.format(oeid))
    for count in ['f_count', 'timestamp_count']:
        notnull = dft[pd.notnull(dft[count])]
        if len(notnull) > 0:
            _dict.update({count:int(dft[pd.notnull(dft[count])][count].iloc[0])})
        else:
            _dict.update({count:None})
    list_of_dicts.append(_dict)
```

```python
def get_osid(oeid):
    return db.lims_query('select ophys_session_id from ophys_experiments where id = {}'.format(oeid))
```

```python
count_comparison = pd.DataFrame(list_of_dicts)
count_comparison['ophys_session_id'] = count_comparison['ophys_experiment_id'].map(get_osid)
```
