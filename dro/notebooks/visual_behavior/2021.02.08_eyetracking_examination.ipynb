## make a plot

```python
t0 = 3000
t_span = 2

pupil_sample_times = [
    2998.5,
    2998.985,
    2999.3,
    3000,
    3001
]

fig = plt.figure(figsize=(15,8))
ax = {}
ax['likely_blinks'] = vbp.placeAxesOnGrid(fig, xspan = [0,1], yspan = [0, 0.2])
ax['pupil_area'] = vbp.placeAxesOnGrid(fig, xspan = [0,1], yspan = [0.3,0.5], sharex=ax['likely_blinks'])

ax['pupil_samples'] = vbp.placeAxesOnGrid(fig, dim = [1,len(pupil_sample_times)], xspan = [0,1], yspan = [0.55, 1])



sdk_eye_data = session.eye_tracking.query('time >= @t0-@t_span and time <= @t0+@t_span')
vba_eye_data = eye_data.ellipse_fits['pupil'].query('time >= @t0-@t_span and time <= @t0+@t_span')

ax['likely_blinks'].plot(
    sdk_eye_data['time'],
    sdk_eye_data['likely_blink'],
    color='orange'
)
ax['likely_blinks'].plot(
    vba_eye_data['time'],
    vba_eye_data['likely_blinks'],
    color='black',
    linestyle='--'
)
ax['likely_blinks'].plot(
    vba_eye_data['time'],
    vba_eye_data['likely_outlier'],
    color='blue',
    linestyle='--'
)

ax['likely_blinks'].set_title('likely blink boolean')


ax['pupil_area'].plot(
    sdk_eye_data['time'],
    sdk_eye_data['pupil_area'],
    color='orange'
)

ax['pupil_area'].plot(
    vba_eye_data['time'],
    vba_eye_data['area'],
    color='black',
    linestyle='--'
)

ax['pupil_area'].plot(
    vba_eye_data['time'],
    vba_eye_data['blink_corrected_area'],
    color='blue',
    linestyle='--'
)

ax['pupil_area'].set_title('pupil area')
ax['likely_blinks'].legend(['sdk output','vba output'])
ax['pupil_area'].set_xlabel('time (s)')

for ii,pupil_sample_time in enumerate(pupil_sample_times):
    text = 'point {}\ntime={:0.2f}'.format(ii, pupil_sample_time)
    ax['likely_blinks'].axvline(pupil_sample_time)
    ax['likely_blinks'].text(pupil_sample_time, 0.5, text)
    
    ax['pupil_area'].axvline(pupil_sample_time)
    ax['pupil_area'].text(pupil_sample_time, 0.5*(sdk_eye_data['pupil_area'].max() - sdk_eye_data['pupil_area'].min()), text)
    
    ax['pupil_samples'][ii].imshow(eye_data.get_annotated_frame(time=pupil_sample_time, pupil=True, eye=True, corneal_reflection=True))
    ax['pupil_samples'][ii].set_title(text)
    ax['pupil_samples'][ii].axis('off')
```
