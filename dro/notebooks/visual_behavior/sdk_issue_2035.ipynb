```python
def append_change_row_to_stimulus_presentations(dataset):
    '''
    adds a column to the dataset.stimulus_presentations dataframe identifying whether a stimulus is a change stimulus
    a change stimulus is defined as the first presentation of a new image_name
    omitted stimuli are ignored
    the first stimulus in the session is ignored
    
    operates inplace
    '''
    # make a copy so we can add a column without affecting original
    df_temp = dataset.stimulus_presentations.copy()
    # add a column with the last_image_name
    df_temp['last_image_name'] = df_temp['image_name'].shift()
    
    # a simple function to identify changes, avoiding omissions
    def is_change(row):
        is_change = (row['image_name'] != row['last_image_name'] 
                     and row['image_name'] != 'omitted' 
                     and row['last_image_name'] != 'omitted'
                     and pd.notnull(row['last_image_name'])
                    )
        return is_change
    
    # append the change column onto the original
    dataset.stimulus_presentations['change'] = df_temp.apply(is_change, axis = 1)
dataset.stimulus_presentations.head(50)
```
