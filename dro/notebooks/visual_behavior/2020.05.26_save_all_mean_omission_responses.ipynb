```python
from visual_behavior.ophys.response_analysis.response_analysis import ResponseAnalysis
import visual_behavior.visualization.ophys.summary_figures as sf
import visual_behavior.database as db
import visual_behavior.data_access.loading as loading
import pandas as pd
import seaborn as sns
from multiprocessing import Pool
import time

import warnings
warnings.filterwarnings("ignore")
```

```python
experiments_table = loading.get_filtered_ophys_experiment_table().reset_index()
```

```python
len(experiments_table)
```

```python
experiment_id = experiments_table.iloc[0]['ophys_experiment_id']
```

```python
def log_mean_omission_responses(experiment_id):
    t0=time.time()
    dataset = loading.get_ophys_dataset(experiment_id)
    analysis = ResponseAnalysis(dataset, overwrite_analysis_files=False, dataframe_format='wide')
    valid_cells = list(dataset.cell_specimen_table.query('valid_roi==True').index.values)
    
    conn = db.Database('visual_behavior_data')
    for cell_specimen_id in valid_cells:

        entry = experiments_table.query('ophys_experiment_id == @experiment_id').iloc[0].to_dict()
        entry.update({'cell_specimen_id':cell_specimen_id})

        t = analysis.omission_response_df.query("cell_specimen_id == @cell_specimen_id")['trace_timestamps'].mean()
        y = analysis.omission_response_df.query("cell_specimen_id == @cell_specimen_id")['trace'].mean()
        entry.update({'t':t, 'y':y})
        db.update_or_create(
            conn['ophys_analysis']['mean_omission_responses'], 
            db.clean_and_timestamp(entry),
            keys_to_check = ['ophys_experiment_id','cell_specimen_id']
        )
    conn.close()
    print('{} took {:0.2f} seconds'.format(experiment_id, time.time()-t0))
```

```python
with Pool(32) as pool:
    pool.map(log_mean_omission_responses, experiments_table['ophys_experiment_id'].unique())
```
