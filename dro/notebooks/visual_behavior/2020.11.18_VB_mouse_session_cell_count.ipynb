```python
import visual_behavior_glm.GLM_analysis_tools as gat
import pandas as pd

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)

%widescreen
```

```python
glm_version = '9a_L2_optimize_by_session'
cell_summary = gat.retrieve_results(search_dict = {'glm_version': glm_version}, results_type='full')
```

```python
cell_summary[['ophys_experiment_id','cell_specimen_id']].sample(5)
```

```python
project_codes = list(cell_summary['project_code'].unique())[:3]
project_codes
```

```python
len(cell_summary)
```

```python
relevant_summary = cell_summary.query('project_code in @project_codes')
```

```python
len(relevant_summary)
```

```python
print('Unique Mice')
for cre_line,specimen_ids in relevant_summary.groupby(['cre_line'])['specimen_id'].unique().items():
    print(cre_line, len(specimen_ids))
```

```python
print('Unique Sessions')
for cre_line,session_ids in relevant_summary.groupby(['cre_line'])['ophys_session_id'].unique().items():
    print(cre_line, len(session_ids))
```

```python
print('Unique Imaging Planes')
for cre_line,experiment_ids in relevant_summary.groupby(['cre_line'])['ophys_experiment_id'].unique().items():
    print(cre_line, len(experiment_ids))
```

```python
print('Unique Cells')
relevant_summary.groupby(['cre_line'])['_id'].count()
```
