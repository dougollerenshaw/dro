```python
from visual_behavior_glm.glm import GLM
import matplotlib.pyplot as plt
```

```python
experiment_id = 989610988

glm = GLM(
    ophys_experiment_id = experiment_id,
    version = '16_events_all_L2_optimize_by_session',
    use_previous_fit=True,
)
```

```python
def plot_cell(ax, glm, cell_id, t0, t1, plot_actual=False):
    
    df = glm.cell_results_df.query('cell_specimen_id == @cell_id').query('fit_trace_timestamps > @t0 and fit_trace_timestamps < @t1')

    if plot_actual:
        ax.plot(df['fit_trace_timestamps'], df['fit_array'])
    ax.plot(df['fit_trace_timestamps'], df['model_prediction'])
```

```python
fig, ax = plt.subplots(figsize=(15,5))
cell_id = 1086591956
t0=1220
t1=1250

glm.load_alternate_model('Full')
plot_cell(ax, glm, cell_id, t0, t1, plot_actual=True)

glm.load_alternate_model('single-visual')
plot_cell(ax, glm, cell_id, t0, t1, plot_actual=False)

glm.load_alternate_model('single-behavioral')
plot_cell(ax, glm, cell_id, t0, t1, plot_actual=False)

ax.legend([
    'actual activity',
    'full model prediction',
    'single-visual prediction',
    'single-behavioral prediction'
])
```
