```python
import pandas as pd
import os
from tqdm import tqdm
import shutil
```

```python
new_image_path = r'//allen/programs/braintv/workgroups/nc-ophys/visual_behavior/lick_detection_validation/'
new_folder_names = [
    'app_annotated_ambiguous',
    'app_annotated_licks',
    'app_annotated_noncontacts',
    'app_annotated_paws'
]
new_image_path
```

```python
full_folder_names = [
    'likely_licks',
    'likely_noncontacts',
    'likely_paws',
    'unlabeled_frames'
]
```

```python
for folder in new_folder_names:
    for image in os.listdir(os.path.join(new_image_path,folder)):
        for existing_folder in full_folder_names:
            if image in os.listdir(os.path.join(new_image_path,existing_folder)):
                print('found {}/{} in {}'.format(folder,image,existing_folder))
                try:
                    os.remove(os.path.join(new_image_path,folder,image))
                    os.remove(os.path.join(new_image_path,existing_folder,image))
                except Exception as e:
                    print(e)
```
