```python
from allensdk.brain_observatory.behavior import trials_processing

from allensdk.brain_observatory.behavior import behavior_ophys_session as bos
from allensdk.internal.api import behavior_ophys_api as boa

import pandas as pd
import numpy as np

%widescreen
```

```python
ophys_experiment_id = 848697604
api = boa.BehaviorOphysLimsApi(ophys_experiment_id)
session = bos.BehaviorOphysSession(api)
```

```python
def recalculate_response_latency(row):
    if len(row['lick_times']>0) and pd.notnull(row['change_time']):
        return row['lick_times'][0] - row['change_time']
session.trials['recalculated_response_latency'] = session.trials.apply(recalculate_response_latency, axis=1)
```

```python
session.trials[session.trials['change_time'].notna()][['change_time','response_time','lick_times','response_latency','recalculated_response_latency']].sample(20,random_state=0)
```
