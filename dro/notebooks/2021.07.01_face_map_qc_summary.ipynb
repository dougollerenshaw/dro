## get experiment_tables, concatenate them
this gets the experiments from the first and second data releases

```python
table_1 = loading.get_filtered_ophys_experiment_table(include_failed_data=False).reset_index()
table_2 = loading.get_second_release_candidates().reset_index()
experiment_table = pd.concat([table_1, table_2])
experiment_table
```

## get the results from the face mask qc process

```python
face_mask_table = pd.read_csv('//allen/programs/braintv/workgroups/nc-ophys/visual_behavior/qc_plots/face_motion_ac_mask/checkplots/2021.06.30_check_plot_qc_summary.csv')
face_mask_table
```

## merge in rig, cre_line, session_type info

```python
merged_table = face_mask_table.merge(
    experiment_table[['ophys_session_id','cre_line','equipment_name','session_type']].drop_duplicates(subset = ['ophys_session_id']),
    on='ophys_session_id',
    how='left'
)
```

```python
merged_table
```

## add a column with simpler rig names: mesoscope or scientifica

```python
def simplify_name(name):
    if pd.isnull(name):
        return None
    elif name.startswith('CAM'):
        return 'scientifica'
    elif name.startswith('MESO'):
        return 'mesoscope'
merged_table['rig_simple_name'] = merged_table['equipment_name'].map(lambda name: simplify_name(name))
```

## What is the distribution of sessions for the two rig types and cre_lines?

```python
merged_table[pd.notnull(merged_table['cre_line'])].groupby(['folder_name','rig_simple_name','cre_line'])['ophys_session_id'].count()
```

## How many sessions do are there for which we have a face mask, but no entry in the experiment table?
There are 368 of them.

```python
extra_face_mask_sessions = list(set(face_mask_table['ophys_session_id']) - set(experiment_table['ophys_session_id'].unique()))
len(extra_face_mask_sessions)
```

```python
merged_table.query('ophys_session_id in @extra_face_mask_sessions').groupby('folder_name')['ophys_session_id'].count()
```

## How many sessions do are there in our experiment table for which we have no face mask?
There are 50 of them

```python
missing_face_mask_sessions = list(set(experiment_table['ophys_session_id'].unique()) - set(face_mask_table['ophys_session_id']))
len(missing_face_mask_sessions)
```

```python
extra_sessions_to_unfail = [ 882968562,  882968564,  887386977,  887386983,  891108771,
        851093283,  856123119,  877057354,  948689200,  978296112,
        987317111, 1059792827, 1076531970, 1076531982, 1107463722]
for session in extra_sessions_to_unfail:
    print(session in missing_face_mask_sessions, session in list(merged_table['ophys_session_id']), session in list(experiment_table['ophys_session_id']))
```

## here are the missing sessions

```python
missing_sessions = experiment_table.query('ophys_session_id in @missing_face_mask_sessions or ophys_session_id in @extra_sessions_to_unfail')
len(missing_sessions['ophys_session_id'].unique())
```

```python
filename= '//allen/programs/braintv/workgroups/nc-ophys/visual_behavior/qc_plots/face_motion_ac_mask/checkplots/2021.07.01_missing_face_mask_inventory.csv'

sessions_needing_masks= experiment_table.query('ophys_session_id in @missing_face_mask_sessions').drop_duplicates(subset = ['ophys_session_id']).set_index('ophys_session_id').reset_index()

sessions_needing_masks.to_csv(filename, index=False)
```
